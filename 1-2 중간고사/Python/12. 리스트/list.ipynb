# 리스트 수정
<hr>

```python
numbers = [2, 4, 6, 8, 10, 1, 3, 5, 7, 9]

numbers[4] = 100
print(numbers)

numbers[2] = "hello"
print(numbers)

numbers[0] = numbers[9] #인덱스 9를 인덱스 0에 대입
print(numbers)

numbers[8] = ['a', 'b', 'c'] #리스트 전체를 형태 유지하며 대입
print(numbers)
```

```python
numbers[:] = [1]
print(numbers)
```

# 리스트 삭제
<hr>

```python
numbers = [2, 4, 6, 8, 10, 1, 3, 5, 7, 9]

a = "goorm"

# 공간까지 삭제
del numbers[4]
print(numbers)

del numbers[:5]
print(numbers)

# 객체 자체를 삭제
del a
print(a) # => 삭제돼서 오류가 발생됨
```
